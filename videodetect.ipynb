```python
import cv2
from ultralytics import YOLO
```

```python
# Load the model
model = YOLO("/content/best.pt")
```

```python
# Input and output video paths
input_video_path = "/content/test_vid.mp4"  # Replace with your video file path
output_video_path = "output_video.avi"  # The output video file
```

```python
# Open the video
cap = cv2.VideoCapture(input_video_path)

# Get video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define VideoWriter to save the output video
fourcc = cv2.VideoWriter_fourcc(*'XVID')  # Codec for .avi format
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

# Process the video
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO inference
    results = model(frame)

    # Annotate the frame with detection results
    annotated_frame = results[0].plot()

    # Write the frame to the output video
    out.write(annotated_frame)

    # Display the frame (optional, may slow down processing in Colab)
    # cv2_imshow(annotated_frame)

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()
```
